```python
# basic libs
import pandas as pd
import numpy as np
```

```python
# fetch dataset
df = pd.read_csv('/home/akash/Datasets/ben10_aliens.csv')
df.head()
```

```python
# EDA & preprocessing
df.describe()
```

```python
df.info()
```

```python
df.shape
```

```python
# missing values
df.isna().sum()
```

```python
# conversion categorical to numeric
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['alien_name'] = encoder.fit_transform(df['alien_name'])
df['species'] = encoder.fit_transform(df['species'])
df['home_planet'] = encoder.fit_transform(df['home_planet'])
df.head()
```

```python
# scale values
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(df)
X_scaled
```

```python
# clustering
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 3)
y_pred = model.fit_predict(X_scaled)
df['cluster'] = y_pred
```

```python
# unique values
df['cluster'].value_counts()
```

```python
df.head(100)
```

```python
# PCA
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pca = PCA(n_components = 3)
pca_data = pca.fit_transform(X_scaled)

plt.figure(figsize=(5, 6))
sns.scatterplot(x = pca_data[:,0], y = pca_data[:,1], hue = df['cluster'], palette = 'Set2', s = 100)
plt.title("KMeans Clustering (2D PCA)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```
